# Normal ve t dağılım

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import scipy
import math
import pandas as pd
import numpy as np
import scipy.stats as stats
%matplotlib inline
```

# 1. Ort = 5, Std = 1, n = 100

```python
x = random.normal(loc=5, scale=1, size=(100))
sns.distplot(random.normal(x), hist=False)

plt.show()
print("Basıklık:" , stats.skew(x))
print("Çarpıklık:" , stats.kurtosis(x))
x
```

# 2. Ort = 1, Std = 2, n = 100

```python
y = random.normal(loc=1, scale=2, size=(100))
sns.distplot(random.normal(y), hist=False)

plt.show()
print("Basıklık:" , stats.skew(y))
print("Çarpıklık:" , stats.kurtosis(y))
y
```
